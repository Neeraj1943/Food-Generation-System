```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.metrics import accuracy_score, classification_report

# Read recipe data from CSV file
csv_file = "/content/drive/MyDrive/epi_r.csv"  # Replace with your actual CSV file
recipe_data = pd.read_csv(csv_file)

# Preprocess the data
le = LabelEncoder()
recipe_data['label'] = le.fit_transform(recipe_data['rating'] > recipe_data['rating'].mean())

# Split the data into training and testing sets
train_data, test_data = train_test_split(recipe_data, test_size=0.2, random_state=42)

# Tokenize and pad the recipe titles
max_words = 1000
tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer.fit_on_texts(train_data['title'])
X_train = tokenizer.texts_to_sequences(train_data['title'])
X_test = tokenizer.texts_to_sequences(test_data['title'])

maxlen = 20  # Adjust as needed
X_train = pad_sequences(X_train, maxlen=maxlen, padding='post')
X_test = pad_sequences(X_test, maxlen=maxlen, padding='post')

# Build the CNN model
embedding_dim = 50  # Adjust as needed
model = Sequential()
model.add(Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=maxlen))
model.add(Conv1D(128, 5, activation='relu'))
model.add(GlobalMaxPooling1D())
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, train_data['label'], epochs=20, batch_size=32, validation_split=0.2)

# Evaluate the model on the test set
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

accuracy = accuracy_score(test_data['label'], y_pred)
print(f"Model Accuracy: {accuracy}")
print(classification_report(test_data['label'], y_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.metrics import accuracy_score, classification_report

# Read recipe data from CSV file
csv_file = "/content/drive/MyDrive/epi_r.csv"  # Replace with your actual CSV file
recipe_data = pd.read_csv(csv_file)

# Preprocess the data
le = LabelEncoder()
recipe_data['label'] = le.fit_transform(recipe_data['rating'] > recipe_data['rating'].mean())

# Split the data into training and testing sets
train_data, test_data = train_test_split(recipe_data, test_size=0.2, random_state=42)

# Tokenize and pad the recipe titles
max_words = 10000  # Increase the number of words to consider
tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer.fit_on_texts(train_data['title'])
X_train = tokenizer.texts_to_sequences(train_data['title'])
X_test = tokenizer.texts_to_sequences(test_data['title'])

maxlen = 30  # Increase the maximum sequence length
X_train = pad_sequences(X_train, maxlen=maxlen, padding='post')
X_test = pad_sequences(X_test, maxlen=maxlen, padding='post')

# Build the CNN model
embedding_dim = 100  # Increase the embedding dimension
model = Sequential()
model.add(Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=maxlen))
model.add(Conv1D(128, 5, activation='relu'))
model.add(GlobalMaxPooling1D())
model.add(Dense(64, activation='relu'))  # Add a dense layer
model.add(Dropout(0.5))  # Add dropout layer for regularization
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, train_data['label'], epochs=20, batch_size=64, validation_split=0.2)

# Evaluate the model on the test set
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

accuracy = accuracy_score(test_data['label'], y_pred)
print(f"Model Accuracy: {accuracy}")
print(classification_report(test_data['label'], y_pred))
```
